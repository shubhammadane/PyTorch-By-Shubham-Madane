```python
# Create an uninitialized tensor with the specified size (2 rows, 3 columns).
# The values in this tensor are not initialized and will contain whatever data is currently in memory.
empty_tensor = torch.empty(2,3)
print(f"Empty Tensor: {empty_tensor}")

# Calculate the mean of the tensor
mean_of_tensor = torch.mean(empty_tensor)
print(f"Mean of the tensor: {mean_of_tensor}")
```

```python
# Check the type of the 'empty_tensor' to confirm it's a PyTorch tensor.
print(f"Type of empty_tensor: {type(empty_tensor)}")

# Display the mean of the tensor, which was calculated in the previous cell.
print(f"Mean of the tensor: {mean_of_tensor}")
```

```python
# Create a tensor of the specified size (2 rows, 3 columns) filled with zeros.
zeros_tensor = torch.zeros(2,3)
print(f"Zeros Tensor: {zeros_tensor}")

# Calculate the mean of the tensor
mean_of_zeros_tensor = torch.mean(zeros_tensor)
print(f"Mean of the zeros tensor: {mean_of_zeros_tensor}")
```

```python
# Create a tensor of the specified size (2 rows, 3 columns) filled with ones.
ones_tensor = torch.ones(2,3)
print(f"Ones Tensor: {ones_tensor}")

# Calculate the mean of the tensor
mean_of_ones_tensor = torch.mean(ones_tensor)
print(f"Mean of the ones tensor: {mean_of_ones_tensor}")
```

```python
# Create a tensor of the specified size (2 rows, 3 columns) filled with random values between 0 and 1.
random_tensor = torch.rand(2,3)
print(f"Random Tensor: {random_tensor}")

# Calculate the mean of the tensor
mean_of_random_tensor = torch.mean(random_tensor)
print(f"Mean of the random tensor: {mean_of_random_tensor}")
```

```python
# Interview Questions & Answers
# Q1. What does `torch.manual_seed(100)` achieve here?
# A1. This function sets the random seed for all CPU operations in PyTorch. By setting a fixed seed (e.g., 100),
#     we ensure that any subsequent random number generation, such as with `torch.rand()`, will produce
#     the exact same sequence of random numbers every time the code is run. This is crucial for ensuring
#     reproducibility in machine learning experiments and debugging.
torch.manual_seed(100)

# Q2. Explain what `torch.rand(2,3)` creates and its typical use cases.
# A2. `torch.rand(2,3)` generates a tensor with 2 rows and 3 columns, filled with random floating-point numbers
#     uniformly distributed over the interval [0, 1). This means the values are greater than or equal to 0
#     and strictly less than 1. It's commonly used for initializing network weights, creating synthetic data for tests,
#     or performing various stochastic operations in deep learning models.
random_seeded_tensor = torch.rand(2,3)

# Q3. How can we verify the content of the created tensor and understand its central tendency?
# A3. We can print the `random_seeded_tensor` to inspect its values and then calculate its mean
#     to get a sense of its average value.
print(f"Random Seeded Tensor: {random_seeded_tensor}")

# Calculate the mean of the tensor
mean_of_seeded_tensor = torch.mean(random_seeded_tensor)
print(f"Mean of the random seeded tensor: {mean_of_seeded_tensor}")
```

```python
# Interview Questions & Answers
# Q1. What is the purpose of `torch.tensor()` and how is it used here?
# A1. `torch.tensor()` is a fundamental PyTorch function used to construct a tensor directly from Python data
#     structures like lists or NumPy arrays. In this example, `torch.tensor([[1,2,3],[4,5,6]])` creates
#     a 2x3 tensor from a nested Python list. Each inner list becomes a row in the tensor.
#     This is useful for creating tensors with specific, predefined values.
custom_tensor = torch.tensor([[1,2,3],[4,5,6]])

# Q2. How can you inspect the content and characteristics of this custom tensor?
# A2. We can print the `custom_tensor` to see its elements. To understand its central tendency,
#     especially for tensors with numerical data, we can calculate its mean.
print(f"Custom Tensor: {custom_tensor}")

# Calculate the mean of the tensor
mean_of_custom_tensor = torch.mean(custom_tensor.float()) # Convert to float for mean calculation
print(f"Mean of the custom tensor: {mean_of_custom_tensor}")
```

```python
# Interview Questions & Answers
# Q1. What does `torch.arange(2, 10, 2)` achieve, and why is it useful?
# A1. `torch.arange()` is used to create a 1-D tensor with a sequence of numbers, similar to Python's `range()`.
#     Specifically, `torch.arange(start, end, step)` generates numbers from `start` (inclusive) up to `end` (exclusive),
#     with a `step` size. Here, it creates a tensor starting at 2, incrementing by 2, and stopping before 10.
#     So, it will produce [2, 4, 6, 8].
#     This is particularly useful for creating sequences for indexing, iteration, generating test data,
#     or defining coordinate axes in various mathematical and machine learning contexts.
sequence_tensor = torch.arange(2,10,2)

# Q2. How can we confirm the contents of this generated sequence and its statistical properties?
# A2. We can print the `sequence_tensor` to see the generated values. To understand its central tendency,
#     we can calculate its mean.
print(f"Sequence Tensor: {sequence_tensor}")

# Calculate the mean of the tensor
mean_of_sequence_tensor = torch.mean(sequence_tensor.float()) # Convert to float for mean calculation
print(f"Mean of the sequence tensor: {mean_of_sequence_tensor}")
```

```python
# Interview Questions & Answers
# Q1. What does `torch.linspace(0, 10, 10)` achieve, and when would you use it?
# A1. `torch.linspace()` generates a 1-D tensor with a specified number of evenly spaced points
#     between a start and end value (inclusive). Here, `torch.linspace(0, 10, 10)` creates 10 points
#     linearly spaced between 0 and 10. The first value will be 0.0 and the last will be 10.0.
#     This is particularly useful when you need a precise number of samples over a defined interval,
#     such as for plotting functions, creating input ranges for models, or for tasks like numerical integration.
linear_spaced_tensor = torch.linspace(0,10,10)

# Q2. How can we inspect the generated sequence and understand its central tendency?
# A2. We can print the `linear_spaced_tensor` to see all the evenly distributed values. To get a summary
#     of its values, calculating its mean provides a central measure.
print(f"Linear Spaced Tensor: {linear_spaced_tensor}")

# Calculate the mean of the tensor
mean_of_linear_spaced_tensor = torch.mean(linear_spaced_tensor)
print(f"Mean of the linear spaced tensor: {mean_of_linear_spaced_tensor}")
```

```python
# Interview Questions & Answers
# Q1. What does `torch.eye(5)` achieve, and when is it typically used?
# A1. `torch.eye(n)` creates a 2-D tensor (a square matrix) with ones on the diagonal and zeros elsewhere.
#     The argument `n` specifies the number of rows and columns, so `torch.eye(5)` creates a 5x5 identity matrix.
#     This function is fundamental in linear algebra and is frequently used in machine learning for tasks such as:
#     - Initializing identity transformations or matrices in neural networks.
#     - Creating masks for specific operations.
#     - Representing the identity element in matrix multiplication.
identity_matrix = torch.eye(5)

# Q2. How can we inspect the content of this identity matrix and understand its statistical properties?
# A2. We can print the `identity_matrix` to visually confirm its structure. Although less common for identity matrices,
#     calculating the mean can provide a simple summary of the overall value distribution.
print(f"Identity Matrix: {identity_matrix}")

# Calculate the mean of the tensor
mean_of_identity_matrix = torch.mean(identity_matrix)
print(f"Mean of the identity matrix: {mean_of_identity_matrix}")
```

```python
# Interview Questions & Answers
# Q1. What does `torch.full((3, 3), 5)` achieve, and when is it typically used?
# A1. `torch.full(size, fill_value)` creates a tensor of a specified `size` where every element is set to `fill_value`.
#     In this case, `torch.full((3, 3), 5)` creates a 2-D tensor (a matrix) with 3 rows and 3 columns, and all 9 elements are set to the value 5.
#     This function is particularly useful for:
#     - Initializing tensors with a constant bias or a specific value.
#     - Creating masks for operations where certain regions need a fixed value.
#     - Setting up initial states for recurrent neural networks or other models that require constant inputs.
#     - Padding tensors to a specific size with a constant value.
full_tensor_3x3 = torch.full((3, 3), 5)

# Q2. How can we inspect the content of this tensor and understand its statistical properties?
# A2. Printing `full_tensor_3x3` allows us to see its 3x3 structure filled with 5s. Calculating the mean of this tensor
#     provides a simple numerical summary. Since all elements are the same, the mean will naturally be equal to the fill value.
print(f"Full 3x3 Tensor: {full_tensor_3x3}")

# Calculate the mean of the tensor
mean_of_full_tensor_3x3 = torch.mean(full_tensor_3x3.float()) # Ensure float for mean calculation
print(f"Mean of the full 3x3 tensor: {mean_of_full_tensor_3x3}")
```
